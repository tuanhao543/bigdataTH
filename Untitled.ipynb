```python
file_path = 'stopwords.txt'
with open(file_path, 'r') as file:
    stpword = file.read()

print(stpword)
```

```python
stpword=stpword.lower()
```

```python
print(stpword)
```

```python
stopwords=set(stopwords.words("english"))
```

```python
from nltk.tokenize import word_tokenize
```

```python
import nltk
nltk.download('punkt')
```

```python
result=word_tokenize(stpword)
```

```python
final=[]
```

```python
for i in result:
    if i not in stopwords:
        final.append(i)
        final.append(" ")
```

```python
"".join(final)
```
